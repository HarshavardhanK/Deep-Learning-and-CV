### Image Blending

```python
img1 = cv.imread("apple.jpg", 1)
img2 = cv.imread("messi.jpg", 1)

plt.imshow(img2)

scale_percent = 60
width = int(img2.shape[1])
height = int(img2.shape[0])

dim = (width, height)

resized_img1 = cv.resize(img1, dim, interpolation = cv.INTER_AREA)
plt.imshow(resized_img1)
```

```python
added = cv.add(img2, resized_img1)
plt.imshow(added)
```
